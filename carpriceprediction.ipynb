```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/carprices.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
final = pd.concat([df,dummies],axis='columns')
final
```

```python
merge = final.drop(['Car Model','BMW X5'],axis='columns')
merge
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = merge.drop('Sell Price($)',axis='columns')
X
```

```python
y = merge['Sell Price($)']
y
```

```python
fit = model.fit(X,y)
```

```python
pre = model.predict([[4500,4,0,3]])
pre
```

```python
model.score(X,y)
```
