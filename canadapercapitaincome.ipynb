```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
val = pd.read_csv("/content/canada_per_capita_income.csv")
val
```

```python
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(val.year, val["per capita income (US$)"], color='green', marker='+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(val[['year']], val[['per capita income (US$)']])
reg.predict([[2005]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
