```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
val = pd.read_csv('/content/hiring.csv')
val
```

```python
import math
median_test_score = math.floor(val['test_score(out of 10)'].mean())
median_test_score
```

```python
val["test_score(out of 10)"] = val["test_score(out of 10)"].fillna(median_test_score)
val
```

```python
val.experience = val.experience.fillna('zero')
```

```python
val
```

```python
from word2number import w2n
val.experience = val.experience.apply(w2n.word_to_num)
val
```

```python
reg = linear_model.LinearRegression()
reg.fit(val[['experience','test_score(out of 10)','interview_score(out of 10)']],val['salary($)'])
```

```python
reg.predict([[2,9,6]])
```
